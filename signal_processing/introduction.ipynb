# Signals

A signal is defined as some value which is dependent on a variable $t$.
Usually, $t$ refers to time, and we refer to the signal as a function, $x(t)$.

Common examples of signals are:
* Amplitude of sound waves
* Voltage on a wire
* Intensity of light

## Classification of signals

There are few ways to classify signals.

### Analog vs Digital

**Definition**: A **continuous-time signal** is a signal $x(t)$ if $t$ is continuous.

For example, the below signal is continuous since the signal is defined for all values of $t$ within the plot.

```python
ts = np.linspace(0, 10)
ts = np.linspace(0, 10)
plt.plot(ts, np.sin(ts))
plt.plot(ts, np.sin(ts))
plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Continuous-time signal")
plt.show()
```

In contrast, the below signal is not continuous as the signal is not defined for some values of $t$.

```python
ts1 = np.linspace(0, 3)
ts2 = np.linspace(6, 10)
plt.plot(ts1, np.sin(ts1), c='b')
plt.plot(ts2, np.sin(ts2), c='b')
plt.xlabel("t")
plt.ylabel("x(t)")
plt.show()
```

An analog signal is a continuous time signal can take on **any value in a continuous range**.
Usually, this is shown as a waveform.

For example, the sine wave below (as shown previously).

```python
plt.plot(ts, np.sin(ts))
plt.plot(ts, np.sin(ts))
plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Analog signal")
plt.show()
```

An digital signal is a continuous time signal can take on **a finite number of distinct values**.

For example, the below is the digital version of the sine wave.

```python
levels = 10
step_size = 2 / levels

ts = np.linspace(0, 10, num=1000)
ts = np.linspace(0, 10, num=1000)
plt.plot(ts, np.round(np.sin(ts) / step_size) * step_size)
plt.plot(ts, np.round(np.sin(ts) / step_size) * step_size)
plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Digital signal")
plt.show()
```

As its name implies, these signals are usually the output of digital devices.

---

**Definition**: A **discrete-time signal** is a signal that is defined only at discrete time points.

A discrete-time signal is usually denoted by $(x_1, \dots, x_n)$, where $x_i$ tells us the value of the signal at time $i$.

Example of discrete-time signal includes:
* the amount of rain per day reported by weather stations
* the daily price of a stock
* discrete sampling of a continuous-time signal

Discrete-time signals can be analog or digital, similar to how we've defined it for the continuous case.
We call these **analog/digital sequences** respectively.
Below are the plots for discrete-time signals.

```python
ts = np.linspace(0, 10)
plt.scatter(ts, np.sin(ts))
plt.scatter(ts, np.sin(ts))
plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Analog sequence")
plt.show()
```

```python
levels = 10
step_size = 2 / levels

ts = np.linspace(0, 10, num=50)
ts = np.linspace(0, 10, num=50)
plt.scatter(ts, np.round(np.sin(ts) / step_size) * step_size)
plt.scatter(ts, np.round(np.sin(ts) / step_size) * step_size)
plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Digital sequence")
plt.show()
```

### Periodic vs Aperiodic

A periodic signal is one which repeats after some time period.
Mathematically, it is defined as:
$$
x(t) = x(t+T) \quad \forall t
$$

The smallest value of $T$ which satisfy the above is called the **fundamental period**, or more succinctly, the period of $x(t)$.

And by definition, $\frac{1}{T}$ when $T$ is in seconds is called the **fundamental frequency**, measured in Hz.

For example, our previous sine wave has a period of $2\pi$ seconds, and a fundamental frequency of $\frac{1}{2\pi}$ Hz.

Trivially, any signal that is not periodic is call **aperiodic**.

```python
ts = np.linspace(0, 20)
plt.plot(ts, np.where(ts < np.pi * 4, np.sin(ts), 0))
plt.plot(ts, np.where(ts < np.pi * 4, np.sin(ts), 0))
plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Aperiodic signal")
plt.show()
```

### Real vs Complex

A **complex signal** is one which is expressed in the form of $a(t) + jb(t)$, where $a(t)$ and $b(t)$ are both real signals, and $j$ is the imaginary unit, defined as $j^2 = -1$.

Similar to complex coordinates, we can express these signals as Cartesian form or exponential form.

The Cartesian form is simply as defined previously:
$$
x(t) = a(t) + jb(t) \\
a(t) = Re\left[ x(t) \right] \\
b(t) = Im\left[ x(t) \right]
$$

And the exponential form is 
$$
x(t) = r(t) e^{j \theta(t)}\\
r(t) = | x(t) |  \\
\theta(t) = \angle x(t)
$$

A **real signal**, thus, is a signal where the complex component $b(t)$ is $0$ (or equivalently, $\theta(t) = k\pi, k \in \mathbb Z$).

### Even vs Odd

An **even** function has the property that:
$$
x(t) = x(-t)
$$

An **odd** function has the property that:
$$
x(t) = -x(-t)
$$


In other words, even functions are symmetrical about $t=0$, while odd functions are symmetrical about $t=0$ if you flip vertically the part where $t > 0$.

For example, $cos(t)$ is even, while $sin(t)$ is odd.

```python
_, axs = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(16, 3))

ts = np.linspace(-6, 6)
axs[0].plot(ts, np.cos(ts))
axs[0].set_xlabel("t")
axs[0].set_ylabel("cos(t)")
axs[0].set_title("Cosine")
axs[0].grid()

axs[1].plot(ts, np.sin(ts))
axs[1].set_xlabel("t")
axs[1].set_ylabel("sin(t)")
axs[1].set_title("Sine")
axs[1].grid()

plt.show()
```

### Deterministic vs Random

A **deterministic** signal is one where each value is of the signal can be determined by some "rule".
This means that future value of the signal can be predicted with complete accuracy.
Hence. deterministic signals carry no information.

In contrast, a **random** signal have some uncertainty about its behaviour.
The future value of $x(t)$ cannot be accurately predicted.
The more random a signal is, the more information it carries.

### Energy vs Power

The **total energy** of a signal is defined as:
$$
E = \int ^ \infty _{-\infty} |x(t)|^2 dt
$$

A signal is an **energy** signal if and only if:
$$
0 < E < \infty
$$

In other words, the area between the signal curve and the $t$-axis is non-zero and finite.

```python
ts = np.linspace(-2, 8, num=100)
ts = np.linspace(-2, 8, num=100)
plt.plot(ts, np.where((ts < np.pi * 2) & (ts > 0), np.sin(ts), 0))
plt.plot(ts, np.where((ts < np.pi * 2) & (ts > 0), np.sin(ts), 0))
plt.fill_between(ts, np.sin(ts), where=(ts < np.pi*2) & (ts > 0))
plt.fill_between(ts, np.sin(ts), where=(ts < np.pi*2) & (ts > 0))
plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Energy signal")
plt.grid()
plt.show()
```

For example, a signal that takes on the value of a sine wave in the range of $(0, 2\pi)$, but 0 everywhere else, is a energy signal.

Its energy is 
$$
E = \int ^\infty _{-\infty} |x(t)|^2 dt 
= \int ^{2\pi} _ 0 \sin^2(t) dt
= \left[ \frac{t}{2} - \frac{\sin(t)}{4} \right]^{2\pi} _ 0 
= \pi
$$

The <span id=power-formula>**average power**</span> of a signal is defined as:
$$
P = \lim _{\tau \to \infty} \frac{1}{2\tau} \int ^ \tau _{-\tau} |x(t)|^2 dt
$$

Graphically, it correlates with the average area between the signal and the $t$-axis as we widen the range of $t$ to infinity.

A signal is an **power** signal if and only if:
$$
0 < P < \infty
$$

For example, a sine wave that is defined from $-\infty$ to $\infty$ is a power signal.

Its average power is 
$$
P = \lim _{\tau \to \infty} \frac{1}{2\tau} \int ^ \tau _{-\tau} |x(t)|^2 dt
= \lim _{\tau \to \infty} \frac{1}{2\tau} \left[ \frac{t}{2} - \frac{\sin(x)}{4} \right]^{\tau} _ {-\tau} \\
= \lim _{\tau \to \infty} \frac{1}{2\tau} \left( \frac{\tau}{2} - \frac{\sin(\tau)}{4} - \frac{-\tau}{2} + \frac{\sin(-\tau)}{4} \right) \\
= \lim _{\tau \to \infty} \frac{1}{2\tau} \left( \tau - \frac{\sin(\tau)}{2} \right) \\
= \lim _{\tau \to \infty} \left(\frac{1}{2} - \frac{\sin(\tau)}{4\tau} \right)
= \frac{1}{2}
$$

Note that for **periodic signals**, we can compute the average power by computing it over its period instead.

_ie_
$$
P = \lim _{\tau \to \infty} \frac{1}{2\tau} \int ^ \tau _{-\tau} |x(t)|^2 dt
= \frac{1}{T} \int ^T _{0} |x(t)|^2 dt
$$

Thus, it follows that **all bounded periodic signals are power signals**.

---

Using the definitions of power and energy, we observe the below properties:
* Energy signals have zero average power
* Power signals have infinite total energy

Also note that it is possible for a signal to neither be a energy nor a power signal.

```python
ts = np.linspace(0, 10)
plt.plot(ts, ts)
plt.plot(ts, ts)
plt.xlabel("t")
plt.ylabel("x(t)")
plt.show()
```

For example, the simple signal above which has an infinite total energy and infinite average power.

###  Unit impulse (Dirac-$\delta$)
$$
\delta(t) = 
\begin{align}
\begin{cases}
\infty \quad & t=0 \\
0\quad & t\neq0 \\
\end{cases}
\end{align}
$$
$$
\int ^ \epsilon _{-\epsilon} \delta(t) dt = 1 \quad \forall \epsilon > 0
$$

In other words, it can be seen as a rectangle signal with unit area, but infinitely thin and tall.

We represent it as an upwards point arrow with length $1$ on the graph.

```python
plt.arrow(0, 0, 0, 1, head_width=0.1, length_includes_head=True)
plt.xlim(-2, 2)
plt.xlabel("t")
plt.title("Unit impulse")
plt.show()
```

#### Property
Even thought the Dirac function may seem like an weird function, it has some useful properties

##### Symmetry

$$
\delta(t) = \delta(-t)
$$

##### Sampling

$$
x(t) \delta(t - \lambda) = x(\lambda) \delta(t - \lambda)
$$

Like the property implies, it means that multiply a signal $x(t)$ with the Dirac function $\delta(\lambda)$ results in the same Dirac function scaled by the value of $x(\lambda)$

```python
fig, axs = plt.subplots(1, 3, sharex=True, sharey=True, figsize=(16, 3))

t, T = 0.5, 1

ts = np.linspace(-3, 3)
axs[0].plot(ts, signal.sinc(ts, T))
axs[0].set_title('x(t) = sinc(t, 1)')
axs[0].grid()

axs[1].arrow(t, 0, 0, 1, head_width=0.1, length_includes_head=True)
axs[1].set_title(f'(), ={t}')
axs[1].grid()


axs[2].arrow(t, 0, 0, signal.sinc(t, T),
             head_width=0.1, length_includes_head=True)
axs[2].plot(ts, signal.sinc(ts, T), ':')
axs[2].set_title('x()(t-)')
axs[2].grid()

plt.xlim(-3, 3)
fig.suptitle('Sampling using Dirac function')
plt.tight_layout()
plt.show()
```

##### Sifting
$$
\int ^\infty _{-\infty} x(t) \delta(t - \lambda) = x(\lambda)
$$

This property follows from the sampling property.
Because multiply it with a Dirac yields the same Dirac function scaled by $x(\lambda)$, the integral becomes simply $x(\lambda)$ as the area of $\delta(t)$ is $1$.

###  Sinusoid

Sinusoid is the collective term for the class of periodic signals in the form of:
$$
x(t) = \mu \cos (\omega_0 t + \phi) = \frac{\mu}{2} \left( e^{j(\omega_0 t + \phi)} + e^{-j(\omega_0 t + \phi)} \right) \\
x(t) = \mu \sin (\omega_0 t + \phi) = \frac{\mu}{j2} \left( e^{j(\omega_0 t + \phi)} - e^{-j(\omega_0 t + \phi)} \right) \\
$$

```python
_, axs = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(16, 3))

axs[0].plot(ts, signal.sin(ts, 1, 0, 1))
axs[0].set_xlabel("t")
axs[0].set_ylabel("sin(t)")
axs[0].set_title("Sine")

axs[1].plot(ts, signal.cos(ts, 1, 0, 1))
axs[1].set_xlabel("t")
axs[1].set_ylabel("cos(t)")
axs[1].set_title("Cosine")
plt.show()
```

The above are the (real) cosine and sine sinusoids.

There is also the complex sinusoid, in the form of
$$
x(t) = \mu e^{j(\omega_0 t + \phi)} = \mu \left( \cos (\omega_0 t + \phi) + j \sin (\omega_0 t + \phi) \right)
$$

Notice that it contains a real component and an imaginary component.
And since $\sin(\theta + \pi/2) = \cos(\theta)$, the 2 components are actually the same wave, but out of phase by $\pi/2$.

Also, by this construction, if we add a sinusoid with its conjugate (_ie_ $\mu e ^{j(-\omega_0t + \phi)}$), then the resultant signal will be **real**, as the imaginary components cancel each other out.

We have the following definitions for the variables:
* $\mu$: magnitude
* $\omega_0$: angular frequency (rad/s)
* $\phi$: phase (rad)
* $\omega_0 t + \phi$: instantaneous phase (rad)
* $f_0 = \frac{\omega_0}{2\pi}$: cyclic frequency (Hz)

## Basic modification of signals

There are a few basic ways we can modify a signal

### Time scaling

We can time-scale the signal by replacing the time variable $t$ by $\alpha t$, $\alpha > 0$.

When $0 < \alpha < 1$, then it is an uniform **expansion** of $x(t)$ along the time axis.

When $\alpha > 1$, then it is an **contraction** of $x(t)$ along the time axis.


```python
ts = np.linspace(0, 10)
ts = np.linspace(0, 10)
plt.plot(ts, np.sin(ts), label='original')
plt.plot(ts, np.sin(ts), label='original')
plt.plot(ts, np.sin(ts * 2), label='alpha = 2')
plt.plot(ts, np.sin(ts * 2), label='alpha = 2')
plt.plot(ts, np.sin(ts * 0.5), label='alpha = 0.5')
plt.plot(ts, np.sin(ts * 0.5), label='alpha = 0.5')
plt.xlabel("t")
plt.ylabel("x(t)")
plt.legend()
plt.title("Time scaling")
plt.show()
```

### Time reversal

For this, we simply replace $t$ with $-t$.

```python
ts = np.linspace(-10, 10)
ts = np.linspace(-10, 10)
plt.plot(ts, np.where(ts < np.pi, np.sin(ts), 0), label='original')
plt.plot(ts, np.where(ts < np.pi, np.sin(ts), 0), label='original')
plt.plot(ts, np.where(-ts < np.pi, np.sin(-ts), 0), label='time reversed')
plt.plot(ts, np.where(-ts < np.pi, np.sin(-ts), 0), label='time reversed')
plt.xlabel("t")
plt.ylabel("x(t)")
plt.legend()
plt.title("Time reversing")
plt.show()
```

### Time shifting

For this, we simply replace $t$ with $t-\beta$.

When $\beta > 0$, then it is **delaying** x(t).

When $\beta < 0$, then it is **advancing** x(t).

```python
ts = np.linspace(0, 10)
plt.plot(ts, np.sin(ts), label='original')
plt.plot(ts, np.sin(ts + 1), label='beta = 1')
plt.plot(ts, np.sin(ts - 1), label='beta = -1')
plt.xlabel("t")
plt.ylabel("x(t)")
plt.legend()
plt.title("Time shifting")
plt.show()
```
