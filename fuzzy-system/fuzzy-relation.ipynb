# Relations

## Crisp Relation
Firstly, we consider the crisp relation, which defines **the presence or absence of a connection between elements of two or more sets**.

For example, we can have a crisp relation between the set of people, which define whether a person is friends with the other.

Or for below, the relationship between the pressure and the temperature of a system.

```python
fig, axs = plt.subplots(1)
fig.suptitle('Pressure against temperature')
axs.plot(range(10) , [1 + 2 * i for i in range(10)], 'r');
axs.set(xlabel='temperature', ylabel='pressure')
axs.set_yticklabels([])
axs.set_xticklabels([]);
```

This follows the theoretical physics, where a temperature would correspond to a specific pressure.

However, in the real world, we know that for a given temperature, there could be a range of possible temperature.

Thus, we introduce the concept of a **fuzzy relation**

## <div id='fuzzy-relation'><div> Fuzzy Relation
A **fuzzy relation** allows relationship between elements of two or more sets to take an infinite number of relationship degrees.

```python
fig, axs = plt.subplots(1)

grade_0 = [0 + 2 * i for i in range(10)]
grade_1 = [1 + 2 * i for i in range(10)]
grade_2 = [3 + 2 * i for i in range(10)]
fig.suptitle('Pressure against temperature')
axs.plot(range(10) , grade_2, 'b', label='grade=0.6');
axs.plot(range(10) , grade_1, 'r', label='grade=1');
axs.plot(range(10) , grade_0, 'g', label='grade=0.8');

axs.fill_between(range(10), grade_0, grade_1, color='g')
axs.fill_between(range(10), grade_1, grade_2, color='b')

axs.set(xlabel='temperature', ylabel='pressure')
axs.set_yticklabels([])
axs.set_xticklabels([])
axs.legend();
```

With the fuzzy relation, a given temperature can give rise to a small range of pressure, each with their respective degree of relationship.

## Representation
To represent a crisp relation, it is simply the subset of the Cartesian product of the crisp sets.

For example, we have {Alice, Bob, Charlie} in $A$ and {Daniel, Eve} in $B$, with Alice and Daniel being friends, and so is Bob and Eve.

We represent the crisp relation as {(Alice, Daniel), (Bob, Eve)}

Or as a table


| $$\chi_R(x_1, x_2)$$| Daniel | Eve  |
| ------- |--:| -:|
| Alice   | 1 | 0 |
| Bob     | 0 | 1 |
| Charlie | 0 | 0 |



For fuzzy set, we need to indicate the strength of the relationship present ($\mu_R(x_1, x_2, \dots, x_n)$)

For example, 


| $$\mu_R(x_1, x_2)$$| Daniel | Eve  |
| ------- |--:| -:|
| Alice   | 1 | 0 |
| Bob     | 0 | 0.8 |
| Charlie | 0.2 | 0 |

The table above would indicate a strong friendship between Alice and Daniel, a decent friendship between Bob and Eve, and a weak friendship between Charlie and Daniel.
For the others, there is no friendship between them.


Of course, the above is just an example to illustrate the point of fuzzy relations, even though the columns and rows are nonsensical fuzzy sets. 
Unless we are determining whether a person that has certain degree of "Alice-ness", "Bob-ness" and "Charlie-ness" would be friends with some others with "Daniel-ness" and "Eve-ness".

## Inference
So given a relationship between two sets $X, Y$, and a variable $x$, we wish to find the resultant variable $y$.



```python
fig, axs = plt.subplots(1)
fig.suptitle('Pressure against temperature')
axs.plot(range(10) , [1 + 2 * i for i in range(10)], 'r');
axs.set(xlabel='temperature', ylabel='pressure');
plt.yticks(range(1, 20, 2));
plt.vlines(4, ymin=0, ymax=9);
plt.hlines(9, xmin=0, xmax=4);
```

From the above graph, it is natural that given a temperature (4), we can find the corresponding pressure at that temperature (9).
This is inference using a crisp number and a crisp relation.


```python
fig, axs = plt.subplots(1)
fig.suptitle('Pressure against temperature')
axs.plot(range(10) , [1 + 2 * i for i in range(10)], 'r');
axs.set(xlabel='temperature', ylabel='pressure');
plt.yticks(range(1, 20, 2));
plt.vlines(4, ymin=0, ymax=9);
plt.vlines(6, ymin=0, ymax=13);
plt.hlines(9, xmin=0, xmax=4);
plt.hlines(13, xmin=0, xmax=6);
```

Now, the above graph shows that, given a range of temperature (4-6), we can find the corresponding range of pressure (9-13).
This is inference using a crisp set and a crisp relation to obtain the corresponding crisp set.


```python
fig, axs = plt.subplots(1)

grade_0 = [0 + 2 * i for i in range(10)]
grade_1 = [1 + 2 * i for i in range(10)]
grade_2 = [3 + 2 * i for i in range(10)]
fig.suptitle('Pressure against temperature')
axs.plot(range(10) , grade_2, 'y', label='grade=0.6');
axs.plot(range(10) , grade_1, 'r', label='grade=1');
axs.plot(range(10) , grade_0, 'g', label='grade=0.8');


axs.fill_between(range(10), grade_0, grade_1, color='g')
axs.fill_between(range(10), grade_1, grade_2, color='y')

axs.set(xlabel='temperature', ylabel='pressure')
plt.yticks(range(1, 20, 2));
plt.vlines(4, ymin=0, ymax=11);
axs.legend();
```

Now we extend the principles to the fuzzy set.
Suppose that we have a crisp set at 4 as per above, when we can obtain that 8 has a grade of 0.8, 9 has a grade of 1 and 10-11 have a grade of 0.6.
Thus, the corresponding fuzzy number is 
$$
0.8/8 + 1/9 + 0.6/10 + 0.6/11
$$

Now, suppose that the membership grade of 4 is not 1.
It would be natural that this lower membership grade should have a diminished output.
A natural way to handle this is to take the minimum of the membership grade of the fuzzy set and the membership grade of the fuzzy relation.

Hence, we get
$$
\mu_B(y) = \min (\mu_A(x), \mu_R(x, y)
$$

```python
fig, axs = plt.subplots(1)

grade_0 = [0 + 2 * i for i in range(10)]
grade_1 = [1 + 2 * i for i in range(10)]
grade_2 = [3 + 2 * i for i in range(10)]
fig.suptitle('Pressure against temperature')
axs.plot(range(10) , grade_2, 'y', label='grade=0.6');
axs.plot(range(10) , grade_1, 'r', label='grade=1');
axs.plot(range(10) , grade_0, 'g', label='grade=0.8');

axs.fill_between(range(10), grade_0, grade_1, color='g')
axs.fill_between(range(10), grade_1, grade_2, color='y')

axs.set(xlabel='temperature', ylabel='pressure')
plt.yticks(range(1, 20, 2));
plt.vlines(4, ymin=0, ymax=10);
plt.vlines(5, ymin=0, ymax=10);
plt.hlines(10, xmin=0, xmax=5);
axs.legend();
```

However, notice that multiple input elements can contribute to the membership grade of the same corresponding element.
For example, temperature = 4 contributes membership grade of 0.6 to the element pressure = 10, while temperature = 5 contributes the membership grade of 0.8 to pressure = 10.

Since we select the largest grade when performing union on the sets, we obtain that 

$$
\mu_B(y) = \max _{x \in X} \min (\mu_A(x), \mu_R(x, y))
$$

Notice that with this formulation, if a element has membership grade of 0 in $A$, the resultant fuzzy set have a membership grade of 0 in $B$.

Thus, suppose that our fuzzy set for temperature is 
$$
A = 0.8/3 + 1/4 + 0.7/5
$$
The corresponding fuzzy set will be
$$
B = 0.8/6 + 0.8/7 + 0.8/8 + 1/9 + 0.7/10 + 0.7/11 + 0.6/12 + 0.6/13
$$
