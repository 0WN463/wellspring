# Derivatives

## Partial derivatives

We defined the derivative for a single-variable function $y = f(x)$ as
$$
\frac{dy}{dx} = f'(x) = \lim _{h \to 0} \frac{f(x + h) - f(x)}{h}
$$

To analyze the derivative for 2-variable functions, we can simply keep one variable constant and differentiate the other:

$$
\begin{gather*}
\frac{\partial f}{\partial x} = f_x(x, y) = \lim _{h \to 0} \frac{f(x + h, y) - f(x, y)}{h} \\
\frac{\partial f}{\partial y} = f_y(x, y) = \lim _{h \to 0} \frac{f(x, y + h) - f(x, y)}{h}
\end{gather*}
$$

Graphically, the represent the gradient of the surface at the point $(x, y)$ in the $x$ and $y$ directions respectively.

### Second partial derivatives

Just like how we can compute the second derivative of single-variable functions, $\frac{d ^2 y}{d x ^2} = f''(x)$, we can also do so for partial derivatives.

Since we can differentiate across either variable, we get the following second partial derivatives:
* $(f_x)_x = f_{xx} = \frac{\partial ^2 f}{\partial x ^2}$
* $(f_x)_y = f_{xy} = \frac{\partial ^2 f}{\partial y \partial x}$
* $(f_y)_x = f_{yx} = \frac{\partial ^2 f}{\partial x \partial y}$
* $(f_y)_y = f_{yy} = \frac{\partial ^2 f}{\partial y^2}$

$\theorem$ Clairuit's Theorem

If $f_{xy}$ and $f_{yx}$ is [continuous](./limits.ipynb#Continuity) in the region around $(a, b)$, then
$$
f_{xy}(a,b) = f_{yx}(a, b)
$$

## Differentiability

For single-variable functions, $f$ is differential at $a$ if and only if $f'(a)$ exists.

However for multi-variable functions, **the existence of the partial derivative don't imply differentiability**.

The intuition is that in a 2-D plot, if the gradient line exists, then we know the rate of change.

But for a 3-D surface plot, knowing the gradient line along the $x$ and $y$ axis does not gives us the rate of change in all directions.
It is possible that the function has a known rate of change along the $x$ and $y$ axis, but that is undefined along another direction.

We need another definition of differentiability for 2-variable functions.

Let $z = f(x,y)$ and $\Delta x, \Delta y$ increments in $x$ and $y$ respectively at $(a, b)$.
Then, it follows that we get this definition:
$$
\Delta z = f(a + \Delta x, b + \Delta y) - f(a,b)
$$

With this, we formally define differentiability for 2-variable functions:

$f(x, y)$ is differentiable if $\Delta z$ can be expressed as:
$$
\Delta z = f_x(a,b) \Delta x + f_y(a,b) \Delta y + \epsilon_1 \Delta x + \epsilon_2 \Delta y
$$

where $\epsilon_1, \epsilon_2$ are functions of $\Delta x, \Delta y$ such that
$$
\lim _{(\Delta x, \Delta y) \to (0, 0)} \epsilon_i(\Delta x, \Delta y)  = 0
$$

By extension, $f$ is differentiable in the region $D \subseteq \mathbb R^2$ if $f$ is differentiable at every point in $D$.

$\example$

Consider the surface: $$z = 2x^2 + xy$$

Then
$$
\begin{gather*}
f_x = 4x + y  \\
f_y = x
\end{gather*}
$$

$$
\begin{align*}
\Delta z &= 2(x + \Delta x)^2 + (x + \Delta x)(y + \Delta y) - (2x^2 + xy) \\ 
&= 2(2\Delta x + (\Delta x) ^2) + \Delta x y + x \Delta y + \Delta x \Delta y \\
&= \Delta x(4 x + y) + x \Delta y + 2(\Delta x) ^2 + \Delta x \Delta y \\
&= f_x \Delta x + f_y \Delta y + (2\Delta x + \Delta y) \Delta x \\
&= f_x \Delta x + f_y \Delta y + \epsilon_1 \Delta x
\end{align*}
$$
where
$$
\epsilon _1 = 2\Delta x + \Delta y
$$

We can easily verify that $\lim _{(\Delta x, \Delta y) \to (0, 0)} \epsilon _ 1 = 0$,
hence the surface is differentiable for any $(x, y)$.

## Linear approximation

$f(x, y)$ is differentiable and $\Delta x, \Delta y$ are small, we get:
$$
\Delta z \approx f_x(a,b) \Delta x + f_y(a,b) \Delta y
$$

Graphically, this means that tangent curves passing through $(a,b)$ must lie on a plane.

We call this plane the **tangent plane**.

We can use this tangent plane to do some approximations for the value of the surface around a point, thus arriving at:
$$
f(a + \Delta x, b + \Delta y) \approx f(a, b) + f_x(a, b) \Delta x + f_y(a, b) \Delta y
$$

$\example$

Consider a cone with radius of 10 cm and height of 25 cm, but each measurement has an error of up to 0.1 cm.
We wish to find the estimate error in the calculated volume of the cone.

$$
\begin{gather*}
f(r, h) = \frac{1}{3} \pi r^2 h \\
f_r = \frac{2}{3} \pi r h\\
f_h = \frac{1}{3} \pi r^2 \\
\end{gather*}
$$

$$
\begin{align*}
\Delta z &\approx f_r(a, b) \Delta r + f_h(a, b) \Delta h  \\
&= \frac{2}{3} \pi (10)(25)(0.1) + \frac{1}{3} \pi (10)^2 (0.1) \\
&= \frac{10\pi}{3}(0.1) \left(50 + 10 \right) \\
&= 20 \pi
\end{align*}
$$

## Relationship between differentiability and partial derivatives

$f$ is differentiable $\implies$ $f_x, f_y$ exists

$f_x, f_y$ are continuous $\implies$ $f$ is differentiable 

$f$ is differentiable $\implies$ $f$ is continuous

Note that the converses do not hold.

## Chain rule

Given $y = f(x)$ and $x = g(t)$, our **chain rule** for single-variable functions is defined as:
$$
\frac{dy}{dt} = \frac{dy}{dx} \frac{dx}{dt}
$$

Similarly, we can extend this to 2-variable functions, suppose $z = f(x,y), x = g(t), y = h(t)$.
$$
\frac{d z}{d t} = \frac{\partial z}{\partial x}\frac{d x}{d t} + \frac{\partial z}{\partial y}\frac{d y}{d t}
$$

This also can be applied if $x,y$ are 2-variable functions too, with $z = f(x,y), x = g(s, t), y = h(s, t)$.
$$
\begin{gather*}
\frac{\partial z}{\partial s} = \frac{\partial z}{\partial x}\frac{\partial x}{\partial s} + \frac{\partial z}{\partial y}\frac{\partial y}{\partial s} \\
\frac{\partial z}{\partial t} = \frac{\partial z}{\partial x}\frac{\partial x}{\partial t} + \frac{\partial z}{\partial y}\frac{\partial y}{\partial t} 
\end{gather*}
$$

## Directional derivatives

The **directional derivative** $D_{\v u} f(x, y)$ is the rate of change of $f$ at the point $(x, y)$ in the direction of $\v u$.

We get its definition by extending from [partial derivatives](#Partial-derivatives).

Given a **unit vector**, $\v u = \mat{a \\ b}$, then 

$$
D_{\v u} f(x, y) = \lim _{h \to 0} \frac{f(x + h a, y + h b) - f(x, y)}{h} 
$$

Note: it is possible for all the directional derivatives to exists, but $f$ is still not differentiable.

In fact, the directional derivative is simply a linear combination of the partial derivatives:
$$
D_{\v u} f(x, y) = f_x(x, y) a + f_y(x, y) b
$$

## Gradient vector

Since the directional derivative is a linear combination of the partial derivative, it makes sense that we can simply our expression using the following identity:
$$
D_{\v u} f (x, y) = \mat{f_x \\ f_y} \cdot \mat{a \\ b} 
= \mat{f_x \\ f_y} \cdot \v u
$$

When we define the following expression, **gradient vector**:
$$
\nabla f(x,y) = \mat{f_x(x, y) \\ f_y(x, y)}
$$

it gives us a convenient expression to find the directional derivative:
$$
D_{\v u} f(x,y) = \nabla f(x, y) \cdot \v u
$$

### Normal to level curve

$\theorem$
Given a function $f(x,y)$ and some point $(a,b)$, if $\nabla f(a, b) \neq \v 0$, 
then $\nabla f(a,b)$ is the normal to the $k$-level curve of $f$ that contains the point $(a, b)$.

<details>
    <summary style="color: blue">$\proof$ (Click to expand)</summary>
    <div style="background: aliceblue">
        $$
        \begin{gather*}
        \text{Let $R(t) = (x(t), y(t)$ be a paramaterization of the $k$-level curve} \\
        f(x, y) = k \\
        \text{Differentiate both sides with respect to t} \\
        f_x x'(t) + f_y y'(t) = 0 \\
        \mat{f_x \\ f_y} \cdot \mat{x'(t) \\ y'(t)} = 0 \\
        \text{Since $\mat{x'(t) \\ y'(t)}$ is the tangent vector} \\
        \text{And $\nabla f = \mat{f_x \\ f_y}$ is orthogonal to it, it follows that it is the normal}\\
        \end{gather*}
        $$
        $QED$
    </div>
</details>

$\example$

Consider the surface 
$$
z = -x^2 - y ^2
$$
and the point $(3, 4)$

The point is contained in the $k$-level curve of 
$$
z = -3^2 - 4^2 = -25
$$

$$
\begin{gather*}
\nabla f = \mat{-2x \\ -2y} \\
\nabla f(3, 4) = \mat{-6 \\ -8} \\
\end{gather*}
$$

Therefore the normal to the $k$-level curve is $\mat{-6 \\ -8}$.

```python
xs = np.linspace(-5, 5, 1000)
plt.plot(xs, np.sqrt(25 - xs**2))
plt.plot(xs, -np.sqrt(25 - xs**2))
plt.title('$k$-level curve')
plt.scatter(3, 4)
plt.arrow(3, 4, -0.3, -0.4, head_width=0.15)
plt.annotate('$\\nabla f(3,4)$', (2, 3))
plt.axis('scaled');
```

Extending this to 3-D, we get:

$\theorem$
Given a function $f(x,y,z)$ and some point $(a,b,c)$, if $\nabla f(a, b, c) \neq \v 0$, 
then $\nabla f(a,b,c)$ is the normal to the $k$-level surface of $f$ that contains the point $(a, b, c)$.

### Direction of fastest/slowest change

Suppose that we are at a point on the surface, which direction should we move so that we move higher (in the $z$-axis) the fastest?

We can rationalize this as being somewhere on a hill, and wondering which direction to walk will be the steepest.

Extending from the definition of the directional derivative and [relationship between dot product and angle](./introduction.ipynb#Dot-product), we get:
$$
\begin{align*}
D_{\v u} f(x,y) &= \nabla f(x, y) \cdot \v u \\
&= \norm{\nabla f(x, y)} \norm{\v u} \cos \theta \\
&= \norm{\nabla f(x, y)} \cos \theta \\
\end{align*}
$$

Which means that the direction along $\nabla f(x,y)$ gives us the maximum rate of change with value $\norm{\nabla f(x,y)}$, while $-\nabla f(x,y)$ gives the minimum.

```python
xs = np.linspace(-5, 5, 1000)
ys = np.linspace(-5, 5, 1000)
xs, ys = np.meshgrid(xs, ys)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(xs, ys, -xs**2-ys**2);
ax.view_init(35, 60)
```

## Surface

We will now formally define a **surface**.

A surface is a subset of $\mathbb R^3$ that is locally the graph of a differentiable function.

We can define it 
* explicitly in the form of 
$$
z = f(x,y)
$$

* parametrically as per [previous chapter](./curves_and_surfaces.ipynb#Surfaces)

* implicitly, usually as the 0-level surface
$$
S = \left\{ (x,y,z): F(x, y, z) = 0 \right\}
$$

$\example$

$F(x, y, z) = x^2 +y^2 + z^2 -1$ defines a sphere with radius 1.

It implicitly defined $z = \sqrt{1 - x^2 - y^2}$ for points with positive $z$, and $z = 1 - \sqrt{1 - x^2 - y^2}$ for points with negative $z$.

## Tangent planes

The **tangent plane** contains all the tangent lines of a surface at that point.

Finding the tangent plane is rather straight forward:
1. Find the $f_x(a,b)$
2. Find the $f_y(a,b)$
3. Find the [cross product](./introduction.ipynb#Cross-product), $\mat{0 \\ 1 \\ f_y(a, b)} \times \mat{1 \\ 0 \\ f_x(a, b)} = \mat{f_x(a, b) \\ f_y(a, b) \\ -1}$

Since 2 lines is enough to define a plane, and both the vectors derives from the partial derivatives are on the plane, we can define the plane using the cross product as the normal to the plane.

## Implicit differentiation

Suppose that we're given a surface defined as $F(x, y, z) = 0$ where $F$ is differentiable.
$F$ may be complicated, and thus difficult to obtain an explicit representation of the function.

An easier way to get the partial derivative is to perform implicit differentiation, with the below theorem.

$\theorem$
Given a surface defined by $F(a,b,c) = 0 $, and $F_z(a,b,c) \neq 0$, 
$$
\begin{gather*}
\frac{\partial z}{\partial x}(a, b, c) = -\frac{F_x(a,b,c)}{F_z(a,b,c)}\\
\frac{\partial z}{\partial y}(a, b, c) = -\frac{F_y(a,b,c)}{F_z(a,b,c)}
\end{gather*}
$$

<details>
    <summary style="color: blue">$\proof$ (Click to expand)</summary>
    <div style="background: aliceblue">
        $$
        \begin{align*}
        F(x, y, z) & = 0  \\
        \text{Differentiate with respect to $x$} \\
        \frac{\partial F}{\partial x} \frac{\partial x}{\partial x}
        +
        \frac{\partial F}{\partial y} \frac{\partial y}{\partial x}
        +
        \frac{\partial F}{\partial z} \frac{\partial z}{\partial x}
        &= 0 \\
        F_x + F_z \frac{\partial z}{\partial x} = 0 \\
        \frac{\partial z}{\partial x} = -\frac{F_x}{F_z} \\
        \text{And similarly for y}
        \end{align*}
        $$
        $QED$
    </div>
</details>

$\example$

Consider the equation
$$
x^3 + y^3 + z^3  + 6xyz = 1
$$

Notice that it is hard to express $z$ with respect to $x$ and $y$.

But we can still obtain the partial derivative by doing implicit differentiation, as per:

$$
\frac{\partial z}{\partial x} = -\frac{F_x}{F_z} 
= -\frac{3x^2 + 6yz}{3z^2 + 6xy}
$$

Note that when $z=f(x,y)$, the partial derivatives are in terms of $x$ and $y$.

But when we perform implicit differentiation, the partial derivatives are in terms of $x, y$ and $z$.

## Extrema

Given some domain $D \subset R^2$:
* A point $(a, b)$ is the **local maximum** if $f(x, y) \leq f(a, b)$ for all points in a disk centered at $(a, b)$ that is in $D$
* A point $(a, b)$ is the **global maximum** if $f(x, y) \leq f(a, b)$ for all points in $D$

It follows that all global maximums are local maximums, but the converse may not hold.

We also define similarly for **local minimums**.

A point that is either a local maximum or local minimum is called a **local extremum**.

If the directional derivatives exists on an extrema point $(a, b)$, then
$$
D_{\v u} f (a, b) = 0
$$
for all directions $\v u$.

### Critical point

Given a differentiable function $f:D\to \mathbb R$, then a **critical point** $(a, b)$ is where 
$$
f_x(a, b) = 0 \quad f_y(a,b) = 0
$$

In other words, they are points where their tangent planes are parallel to the $xy$ axis.

Note that is it possible for local extrema to be non-critical points, when the point is not in the interior of $D$.

$\example$

Consider the function $f$ defined for $x \geq 0$:
$$
f(x, y) = x
$$

The local minimum is along $x = 0$, but none of the points there are critical points.

In fact, we can reason all of these in 2-D.

A point on the curve may be the lowest/highest, but it may not be stationary point if it lies on the edge of the domain.

Similarly, a stationary point may not an extrema, _ie_ in the case of points of inflexions.

### Saddle point

A **saddle point** $(a, b)$:
1. is a **critical point** 
2. is in a region that contains points above it and below it

More formally, (2) means that every open disk centered at $(a, b)$ contains points $(x, y) \in D$ where $f(x,y) < f(a,b)$ and points $(x, y) \in D$ where $f(x,y) > f(a, b)$

For a smooth function, every critical point is either a local maximum, local minimum or saddle point.

### Global extrema

In general, global maximums/minimums may not exists for a function.

But if we restrict our domain to a **closed and bounded** region, then the global maximum/minimum always exists.

A set is **bounded** if it is contained in some disk.
In other words, it doesn't extend indefinitely.

A set $R \subseteq \mathbb R^2$ is **closed** if it contains all its **boundary points**.

A **boundary point** is a point of $R$ such that every disk centered on it contains points in $R$ and points outside of $R$.

#### Finding global extrema

It is straightforward to find the global extrema base on the different type of points we've discussed above.

We know that the global extrema must exist either:
1. As a critical point on the interior of $D$
2. As an extrema on the boundary of $D$

So we simply find the maximum/minimum within the 2 sets of points and select the larger/smaller.

```python
ax = plt.axes(projection='3d')
ax.plot_surface(xs, ys, xs**2 - 2*xs*ys + 2 * ys);
```

## Lagrange multiplier

Suppose that we are interested in the extrema along some constraint domain, instead of across the whole domain.

For example, we are walking along some trail on a hill, and wish to find where is the highest point along this trail.
(As compared to finding the highest point of the hill using global extrema)

The way to find the extrema is using the **Method of Lagrange Multipliers**.

We first observe the constraint curve $C$ together with the level curves of the surface.

As we move along the constraint curve $C$, we are either going up the level curves or down the level curves.

What is important is that **when we reach a maximum along $C$, we will move down or along the level curve**.

Notice that it means that $C$ and the level curve must share the same tangent line at the extrema.

It also means that their gradient vector must be multiples of each other, _ie_
$$
\nabla f(a, b) = \lambda \nabla g(a, b)
$$

The constant $\lambda$ is known as the **Lagrange Multiplier**.

Hence, the method of Lagrange multipliers is as follows:
Given some function $f(x,y)$ to find the extremas, and some constraint $g(x,y) = k$

1. Find all points where $\nabla f(x, y) = \lambda \nabla g(x,y)$ and $g(x, y) = k$
2. Evaluate $f$ at all the points in (1)
    * the largest is the maximum
    * the lowest is the minimum

Note that this is only guaranteed to work if $C$ is closed and bounded.

$\example$

Suppose that we are given a fence of length 8 and wish to surround the largest rectangular area.

Let $w$ and $l$ be the length and width of the rectangular area respectively.

We getting the following constraint:

$$
g(w,l) = 2w + 2l = 8
$$

And we wish to maximize the following function:
$$
A = f(w, l) = w l
$$

Hence, to solve with the Lagrange multiplier method, we find:
$$
\begin{gather}
\nabla g = \mat{2 \\ 2} \\
\nabla f = \mat{l \\ w} \\
\Rightarrow \mat{l \\ w} = \lambda \mat{2 \\ 2} \\
\Rightarrow \begin{cases} l = 2 \lambda \\ w = 2 \lambda \end{cases}
\end{gather}
$$

Together with
$$
2w + 2l = 8
$$

we can solve simultaneously to get 
$$
\begin{gather}
\lambda = 1 \\
w = l = 2
\end{gather}
$$

As a sanity check, we can use our simpler calculus method to solve this problem:

$$
2w + 2l = 8 \Rightarrow l = 4 - w
$$

$$
\begin{align*}
f(w,l) &= wl = w(4-w) = f(w) \\
f'(w) &= 4 - 2w
\end{align*}
$$

At maximum area, $f'(w) = 0$, thus solving:
$$
4 - 2 w = 0 \Rightarrow w = l = 2
$$

which is the same as using our Lagrange multiplier method.

```python
ws = np.linspace(0, 4)
ls = np.linspace(0, 4)
xs, ys = np.meshgrid(ws, ls)
ax = plt.axes(projection='3d')
ax.plot3D(ws, 4-ws, 0);
ax.plot_surface(xs, ys, xs * ys);
ax.view_init(25, -30)
ax.set_xlabel('$w$')
ax.set_ylabel('$l$')
ax.set_zlabel('Area')
plt.show()
```

Visually, we have this surface and we wish to reach the highest point on the surface, while being restricted to "walking along the constraint below".

This problem is overkill to use this method, but it comes in handy when we are dealing with a problem with more than 2 variables, as per below:

$\example$

Suppose that we wish have $12m^2$ of cardboard and wish to make a box without lid that has the largest volume.

Let $h, w, l$ be the height, width and length of the box respectively.

$$
\begin{gather}
g(h, w, l ) = 2hl + 2 hw + wl = 12\\
f(h, w, l) = hwl \\
\nabla g = \mat{2l + 2w\\ 2h + l \\2h + w} \\
\nabla f = \mat{wl \\ hl \\ hw}\\
2l + 2w = \lambda (wl)\\
2h + l = \lambda (hl)\\
2h + w = \lambda (hw)\\
2hl + 2hw = 2hw + wl = 2hl + wl\\
w = 2h \\
w = l \\
w^2 + w^2 + w^2 = 12\\
w = 2, l = 2, h = 1\\
\end{gather}
$$
