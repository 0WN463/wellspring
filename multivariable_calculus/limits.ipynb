# Limits

For 1-variable functions, we define the limit as
$$
\lim _{x \to a} f(x) = L
$$

which means that the value of $f(x)$ approaches $L$ as $x$ approaches $a$.

More formally, the condition means that:

If $\forall \epsilon > 0$, $\exists \delta > 0$ such that when $0 < |x - a| < \delta$, $|f(x) - L| < \epsilon$.

Note that we can approach the limit in either the positive or negative $x$ direction.
If the limit we get from the two directions are different, then the limit doesn't exists.

$\example$ <div id='eg-sinc'/>

Consider the limit
$$
\lim _{x \to 0} \frac{\sin x}{x}
$$

By L'hopital's rule, we know that 
$$
\lim _{x \to 0} \frac{\sin x}{x} 
= \lim _{x \to 0} \frac{\cos x}{1} 
$$

If we approach from $x > 0$, we see that $\lim _{x \to 0 ^+} \frac{\cos x}{1} = 1$

Similarly, if we approach from $x < 0$, we see that $\lim _{x \to 0 ^-} \frac{\cos x}{1} = 1$

Hence, the limit is $1$.

```python
xs = np.linspace(-1, 1)
plt.title("$\sin x / x$")
plt.scatter(xs, np.sin(xs) / xs);
```

Note that $\frac{\sin x}{x}$ is not defined at $0$, but the limit exists.

Consider the limit
$$
\lim _{x \to 0} \frac{x}{|x|}
$$

If we approach from $x > 0$, we see that $\lim _{x \to 0^+} \frac{x}{|x|} = 1$.

Conversely, if we approach from $x < 0$, we see that $\lim _{x \to 0^-} \frac{x}{|x|} = -1$.

Since $1 \neq -1$, it must mean that the limit doesn't exists.

```python
plt.title("$x / |x|$")
plt.scatter(xs, xs / np.abs(xs));
```

We see that it is discontinuous at 0.

Even though we are dividing 0 by 0, or dividing infinity by infinity, it is possible that the numerator scales at a constant rate with respect to the denominator, giving us a constant for the limit.

In general, there are a few cases for the limit (assuming $x \to 0$):
* no denominator (finite limit)
    * $\lim _{x \to 0} 3x = 0$
* denominator goes to 0 at a fixed rate compared to numerator (finite limit)
    * $\lim _{x \to 0} \frac{3x}{x} = 3$
* denominator goes to 0 faster than numerator (infinite limit)
    * $\lim _{x \to 0} \frac{1}{x^2} = \infty$
    * $\lim _{x \to 0} -\frac{1}{x^2} = -\infty$
    * $\lim _{x \to 0} \frac{x}{x^3} = \infty$
* limit is not defined
    * $\lim _{x \to 0} \frac{1}{x}$

Extending this definition to 2-variables, we get:

$$
\lim _{(x, y) \to (a, b)} f(x, y) = L
$$

If $\forall \epsilon > 0$, $\exists \delta > 0$ such that when $0 < \sqrt{(x-a)^2 + (y-b)^2} < \delta$, $|f(x, y) - L| < \epsilon$.

Note that for 1-variable, we can approach the point in 2 direction, either the positive x or negative x direction.

But for 2-variables, we can approach the point in infinitely many directions, which is any linear ratio in the $x$ and $y$ direction.

To prove that the limit doesn't exists, we once again just need to find 2 directions that gives us different values for the limit.

$\example$

Consider the limit
$$
\lim _{(x, y) \to (1, 0)} \frac{y}{x + y - 1}
$$

If we approach from $y = 0, x > 1$, we see that $\lim _{(x-1) \to 0^+} \frac{0}{x-1} = \infty$.

Conversely, if we approach from $y=0, x < 1$, we see that $\lim _{(x-1) \to 0^-} \frac{0}{x-1} = -\infty$.

Thus the limit doesn't exists.

Note:

If we approach from $x = 1, y > 0$, we see that $\lim _{y \to 0^+} \frac{y}{y} = 1$.

Conversely, if we approach from $x=1, y < 0$, we see that $\lim _{y \to 0^-} \frac{y}{y} = 1$.

Thus, even if we found 2 directions that gives the same limit, **it does not mean that the limit exists**.
It has to be the same for all paths to the point.

To prove the existence of a limit, we need to either:
* Deduce from known limits of functions and using properties of limits
* use the **Squeeze theorem**.

$\theorem$
Let $\lim _{(x,y) \to (a,b)} f(x) = L_1$, $\lim _{(x,y) \to (a,b)} g(x) = L_2$

Then:
$$
\lim _{(x,y) \to (a,b)} \left( f(x) \pm g(x) \right) = L_1 \pm L_2
$$
$$
\lim _{(x,y) \to (a,b)} \left( f(x) g(x) \right) = L_1 L_2
$$
$$
\lim _{(x,y) \to (a,b)} \left( f(x) / g(x) \right) = L_1 / L_2 \quad \text{if } L_2 \neq 0
$$

$\theorem$ **Squeeze Theorem**

Suppose that $g(x,y) \leq f(x,y) \leq h(x,y)$ for all $(x,y)$ in a disk region centered at $(a,b)$ (possibly excluding $(a,b)$.

If
$$
\lim _{(x, y) \to (a,b)} g(x,y) = L = \lim _{(x,y) \to (a,b)} h(x, y)
$$

Then the limit of $f(x, y)$ at $(a,b)$ is $L$ too.

In other words, if we can find a function $g$ that is always below $f$ around $(a,b)$ and a function $h$ that is always above $f$ around $(a,b)$, and somehow we know that their limits are the same at $(a, b)$, then we can say that it is also the limit for $f$ there.

$\example$

Consider the limit
$$
\lim _{(x, y) \to (0, 0)} \frac{3xy^2}{x^2 + y^2}
$$
$$
\begin{align*}
f(x, y) &= \frac{3xy^2}{x^2 + y^2}\\
|f(x, y)| &= \left|\frac{3xy^2}{x^2 + y^2} \right | \\
&= \left| x \right | \frac{3y^2}{x^2 + y^2} \\
&\leq \left| x \right | \left( \frac{3y^2}{x^2 + y^2} + \frac{3x^2}{x^2 + y^2} \right) \\
&= \left| x \right | \left( \frac{3(x^2 + y^2)}{x^2 + y^2} \right) \\
&= 3 \left| x \right |  \\
\therefore -3|x| \leq & f(x ,y) \leq 3|x|
\end{align*}
$$

We know that $\lim _{(x,y) \to (0, 0)} 3|x| = 0$ and similarly for $-3|x|$, hence the limit of $f(x,y)$ is $0$

```python
ss = np.linspace(-1, 1, 1000)
ts = np.linspace(-1, 1, 1000)
xs, ys = np.meshgrid(ss, ts)
```

```python
g = -3 * np.abs(xs)
f = 3 * xs * ys ** 2 / (xs ** 2 + ys ** 2)
h = 3 * np.abs(xs)
```

```python
def mask(xs, ys, zs):
    xs, ys = xs.copy(), ys.copy()
    xs[np.isnan(zs)] = np.nan
    ys[np.isnan(zs)] = np.nan
    return xs, ys, zs
```

```python
fig = plt.figure(figsize=(20, 5))
axs = [fig.add_subplot(1, c, 1, projection='3d') for c in [1, 2]]

plot = axs[0].plot_surface(*mask(xs, ys, g), label='g(x, y)');
plot._edgecolors2d = plot._edgecolor3d
plot._facecolors2d = plot._facecolor3d

plot = axs[0].plot_surface(*mask(xs, ys, f), label='f(x, y)');
plot._edgecolors2d = plot._edgecolor3d
plot._facecolors2d = plot._facecolor3d

plot = axs[1].plot_surface(*mask(xs, ys, h), label='h(x, y)');
plot._edgecolors2d = plot._edgecolor3d
plot._facecolors2d = plot._facecolor3d

plot = axs[1].plot_surface(*mask(xs, ys, f), label='f(x, y)');
plot._edgecolors2d = plot._edgecolor3d
plot._facecolors2d = plot._facecolor3d

for ax in axs:
    ax.legend()
    ax.view_init(5, 90)
```

We can check in that $g(x,y) \leq f(x,y) \leq h(x,y)$ in the region around $(0, 0)$.

It might be easier to notice this if we see that both $f(x,y) - g(x,y)$ and $h(x,y) - f(x,y)$ is always positive in that region.

```python
fig = plt.figure(figsize=(20, 5))
axs = fig.subplots(1, 2)

plot = axs[0].pcolormesh(xs, ys, f - g)
fig.colorbar(plot, ax=axs[0])

plot = axs[1].pcolormesh(xs, ys, h - f)
fig.colorbar(plot, ax=axs[1])
axs[0].set_title('f(x, y) - g(x,y)')
axs[1].set_title('h(x, y) - f(x,y)');
```

# Continuity

We say that $f$ is **continuous** at $(a, b)$ if 
$$
\lim _{(x,y) \to (a,b)} f(x,y) = f(a, b)
$$

If $f$ is not continuous at $(a,b)$, we call it a **discontinuity** at $(a, b)$.

$f$ is continuous in a region $D \subseteq \mathbb R^2$ if $f$ is continuous at every point in $D$.

For our [previous example](#eg-sinc),$\frac{\sin x}{x}$ is discontinuous/undefined at $x = 0$.
Which is [why we augment it](../signal_processing/introduction.ipynb#Sinc-(Sine-cardinal)) with $f(x) = 1$ at $x = 0$, so that we have a continuous function.

$\theorem$
Let $f, g$ be 2 continuous functions at $(a, b)$

Then at $(a,b)$:

$f \pm g$ is continuous

$f(x, y) g(x, y)$ is continuous

$\frac{f}{g}$ is continuous if $g(a,b) \neq 0$

$\theorem$
Let $f$ be continuous functions at $(a, b)$ and $g$ be a continuous function at $f(a, b)$

$g \cdot f$ is continuous at $(a, b)$

```python
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d.axes3d import Axes3D
from mpl_toolkits.mplot3d.proj3d import proj_transform

# https://gist.github.com/WetHat/1d6cd0f7309535311a539b42cccca89c

class Arrow3D(FancyArrowPatch):

    def __init__(self, x, y, z, dx, dy, dz, *args, **kwargs):
        super().__init__((0, 0), (0, 0), *args, **kwargs)
        self._xyz = (x, y, z)
        self._dxdydz = (dx, dy, dz)

    def draw(self, renderer):
        x1, y1, z1 = self._xyz
        dx, dy, dz = self._dxdydz
        x2, y2, z2 = (x1 + dx, y1 + dy, z1 + dz)

        xs, ys, zs = proj_transform((x1, x2), (y1, y2), (z1, z2), self.axes.M)
        self.set_positions((xs[0], ys[0]), (xs[1], ys[1]))
        super().draw(renderer)
        
    def do_3d_projection(self, renderer=None):
        x1, y1, z1 = self._xyz
        dx, dy, dz = self._dxdydz
        x2, y2, z2 = (x1 + dx, y1 + dy, z1 + dz)

        xs, ys, zs = proj_transform((x1, x2), (y1, y2), (z1, z2), self.axes.M)
        self.set_positions((xs[0], ys[0]), (xs[1], ys[1]))

        return np.min(zs) 
def _arrow3D(ax, x, y, z, dx, dy, dz, *args, **kwargs):
    '''Add an 3d arrow to an `Axes3D` instance.'''

    arrow = Arrow3D(x, y, z, dx, dy, dz, *args, **kwargs)
    ax.add_artist(arrow)


setattr(Axes3D, 'arrow3D', _arrow3D)
```

```python
ax = plt.axes(projection='3d')
ax.set_title("$R(t) = (2t, 5t, -3t)$")
ax.plot3D(2 * ts, 5 * ts, -3 * ts);
ax.arrow3D(1, 1, 1, 5, 5, 5, mutation_scale=20)
ax.view_init(20, 150)
```

$\theorem$
The length of $C$ between $R(a)$ and $R(b)$ is:
$$
    \int ^b _a \sqrt{f'(t)^2 + g'(t)^2 + h'(t)^2 } dt
$$

PROOF

$\example$

Suppose that we wish to find the length of
$$
R(t) = (\sin t, \cos t, 0)
$$
from $t = 0$ to $t = 2 \pi$.

Astute readers would notice that this is a full circle with a radius of 1.

```python
ts = np.linspace(0, 2 * np.pi)
ax = plt.axes(projection='3d')
ax.plot3D(np.sin(ts), np.cos(ts), 0);
ax.arrow3D(0, -1, 0, 1, 0, 0, mutation_scale=10);
```

Since the $z$ axis is always $0$, we can plot this on a 2-D plot.

```python
plt.plot(np.sin(ts), np.cos(ts))
plt.axis('scaled');
```

Using our formula, we get
$$
\begin{align*}
\int ^{2\pi} _0 \sqrt {(\cos t)^2 + (-\sin t)^2 + 0^2} dt 
&= \int ^{2\pi} _0 \sqrt {1} dt \\
&= 2 \pi
\end{align*}
$$

And this follows our known formula for the circumference of a circle, $2 \pi r = 2 \pi (1) = 2 \pi$

Suppose we wish to find the velocity of the curve at a point.

Suppose that $f, g, h$ are all differentiable.
The tangent vector to the curve $C$ at the point $R(a)$ is:
$$
\left( R(a), \mat{f'(a) \\ g'(a) \\ h'(a)} \right)
$$

Note that in this case, the parameterization of the curve is important as it will affect **tangent vector** (but parameterization has no effect on the **tangent line**)

# Surfaces

A function can also be defined by 2 variables
$$
R(s, t) = (f(s, t), g(s, t), h(s, t))
$$

To visualize a function in $\mathbb R^2$ with 1 variable, we can plot the 2-D graph of the curve.

We can do similarly for a function in $\mathbb R^3$ with 2 variable, where we plot the **surface**.

```python
ss = np.linspace(-5, 5, 1000)
ts = np.linspace(-5, 5, 1000)
ss, ts = np.meshgrid(ss, ts)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(ss, ts, 3 * ss * ts **2 / (ss ** 2 + ts ** 2), alpha=0.25);
ax.plot_surface(ss, ts, 3 * ss, alpha=0.5);
#ax.plot_surface(ss, ts, -3 * ss);
ax.view_init(5, 55)
```

## Level curve

It is often easier to visualize the graph "slice-by-slice".

The **$k$-level curve** which represents a graph slice is defined as:
$$
\{ (x, y) \in \mathbb R^2: f(x,y) = k \}
$$

This is the same idea as contours lines on a map.

```python
ax = plt.axes(projection='3d')
ax.contour(ss, ts, -ss**2-ts**2, levels=np.arange(-50, 0, 3));
```

Each vertical slice is a $k$-level curve.

## Cylinder

For this module, we have a much broader definition of a **cylinder**.

A surface is a cylinder if there exists some plane $P$ such that all planes parallel to $P$ intersects the surface to create the same curve.

Any equation in $x, y, z$ where one of the variable is missing will create a cylinder.

## Quadric surface

A quadric surface is of the form:
$$
Ax^2 + By^2 + Cz^2 + D xy + E yz + F xz + G x + H y + I z + J = 0
$$

Excluding cylinders, there are 6 basic quadric surfaces

| Name | Equation |
| --- | --- |
| Elliptic paraboloid | $$\frac{x^2}{a^2} + \frac{y^2}{b^2} = \frac{z}{c}$$ |
| Hyperbolic paraboloid |$$\frac{x^2}{a^2} - \frac{y^2}{b^2} = \frac{z}{c}$$ | 
| Ellipsoid |$$\frac{x^2}{a^2} + \frac{y^2}{b^2} + \frac{z^2}{c^2} = 1$$ | 
| Ellipsoid cone |$$\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = 0$$ | 
| Hyperboloid of 1 sheet |$$\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = 1$$ | 
| Hyperboloid of 2 sheet |$$\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = -1$$ | 

```python
xs, ys = ss, ts
```

```python
def mask(xs, ys, zs):
    xs, ys = xs.copy(), ys.copy()
    xs[np.isnan(zs)] = np.nan
    ys[np.isnan(zs)] = np.nan
    return xs, ys, zs
```

```python
with np.errstate(invalid='ignore'):
    zs = xs**2 + ys **2
    
ax = plt.axes(projection='3d')
ax.set_title('Elliptic paraboloid')
ax.plot_surface(*mask(xs, ys, zs));
```

```python
with np.errstate(invalid='ignore'):
    zs = xs**2 - ys **2
    
ax = plt.axes(projection='3d')
ax.set_title('Hyperbolic paraboloid')
ax.plot_surface(*mask(xs, ys, zs));
```

```python
with np.errstate(invalid='ignore'):
    zs = np.sqrt(1- xs**2 - ys **2)
    
ax = plt.axes(projection='3d')
ax.set_title('Ellipsoid')
ax.plot_surface(*mask(xs, ys, zs));
ax.plot_surface(*mask(xs, ys, -zs));
```

This is a sphere when $a = b = c$.

```python
zs = np.sqrt(xs**2 + ys **2)
    
ax = plt.axes(projection='3d')
ax.set_title('Elliptic cone')
ax.plot_surface(*mask(xs, ys, zs));
ax.plot_surface(*mask(xs, ys, -zs));
ax.view_init(10, 35)
```

```python
with np.errstate(invalid='ignore'):
    zs = np.sqrt(xs**2 + ys **2 - 1)
    
ax = plt.axes(projection='3d')
ax.set_title('Hyperboloid of 1 sheet')
ax.plot_surface(*mask(xs, ys, zs));
ax.plot_surface(*mask(xs, ys, -zs));
ax.view_init(10, 35)
```

```python
zs = np.sqrt(xs**2 + ys **2 + 1)
    
ax = plt.axes(projection='3d')
ax.set_title('Hyperboloid of 2 sheet')
ax.plot_surface(*mask(xs, ys, zs));
ax.plot_surface(*mask(xs, ys, -zs));
ax.view_init(10, 35)
```

# Function of 3 variables

By extension, a function can also be defined by 3 variables
$$
R(s, t, u) = (f(s, t, u), g(s, t, u), h(s, t, u))
$$

However, these are rather hard to visualize as we will need to plot them in 4-D.

Instead, we reason about them by looking at their $k$-level surfaces instead.

## $k$-level surface

Extending the definition, we can see that a "slice" of a function with 3 variable is a surface, defined as:

$$
\{ (x, y, z) \in \mathbb R^3: f(x,y, z) = k \}
$$


This provides us some visual aid in reasoning with these functions.
